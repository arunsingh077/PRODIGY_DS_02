The sinking of the Titanic on April 15, 1912, during its inaugural journey, is renowned as one of the most notorious maritime disasters.
```python
#importing data

import pandas as pd
import matplotlib.pyplot as plt
csv_file = r'train.csv'

df=pd.read_csv(csv_file)
df.dtypes
```

```python
# Separate the data into different series
PassengerID = df['PassengerId']
Survived = df['Survived']
Name = df['Name']
Sex = df['Sex']
Age = df['Age']
```

## DATA CLEANING: 

```python
#removing unwanted columns
columns_to_remove = ['SibSp', 'Parch', 'Ticket', 'Fare','Cabin','Embarked']
df = df.drop(columns=columns_to_remove, errors='ignore')
```

```python
print(df)
```

```python
df.head()
```

```python
# Check for missing values
print("Missing values before cleaning:")
print(df.isnull().sum())
```

```python
#dropping rows with missing values
df.dropna(subset=['Age'], inplace=True)
```

```python
# Check missing values after cleaning
print("\nMissing values after cleaning:")
print(df.isnull().sum())
print("\nCleaned DataFrame:")
print(df)
```

## EDA(Exploratory Data Analysis): 

```python
#Importing necessary libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Descriptive Statistics
print(df.describe())
```

```python
# Countplot of Survivors
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df)
plt.title('Count of Survived and Not Survived')
plt.show()
```

```python
# Distribution of 'Age'
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(x='Age', hue='Survived', data=df, fill=True, common_norm=False)
plt.title('Age Distribution by Survival Status')
plt.show()
```

### Inference:  If there's a concentration of younger passengers among survivors, it could suggest that priority was given to children during the evacuation. Conversely, if there's a concentration of older passengers among non-survivors, it might indicate challenges faced by the elderly.

```python

# Survival Rate by Sex
plt.figure(figsize=(8, 5))
sns.barplot(x='Sex', y='Survived', data=df, ci=None)
plt.title('Survival Rate by Sex')
plt.show()
```

### This shows a significant difference in survival rates between males and females. This aligns with historical accounts and the prioritization of women and children during the evacuation.

```python
# Countplot of Survival by Sex
plt.figure(figsize=(8, 5))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Count of Survival by Sex')
plt.show()
```

### Inference: The analysis suggests that gender could be a strong predictor of survival on the Titanic. Females may have a higher likelihood of survival compared to males.
